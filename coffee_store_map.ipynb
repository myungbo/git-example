# 全台咖啡店地圖

```python
# 中文店名有亂碼問題，在這裡進行轉碼
def parse_zhch(s):
    return str(str(s).encode('ascii' , 'xmlcharrefreplace'))[2:-1]

# 定義繪製地圖中心位置和縮放大小
m = folium.Map((23.973875, 120.982024),zoom_start=7)

# Cama(RED)
csvFileToRead = open('C:\\Users\\NB\\notebooks\\hello-git\\camalist.csv', 'r', encoding='utf-8_sig')
csvDataToRead = csv.reader(csvFileToRead)
dataList = list(csvDataToRead)
csvFileToRead.close()

i = 0
while i <=127:
    i = i + 1
    mrt = (dataList[i][5],dataList[i][6])
    name = dataList[i][1]
    folium.Circle(
        location=mrt,
        radius=100,
        popup=parse_zhch(name),
        color='red',
        fill=False,
        fill_color='red',
        fill_opacity=0.05
    ).add_to(m)

# Louisa(BLUE)
csvFileToRead = open('C:\\Users\\NB\\notebooks\\hello-git\\louisalist.csv', 'r', encoding='utf-8_sig')
csvDataToRead = csv.reader(csvFileToRead)
dataList = list(csvDataToRead)
csvFileToRead.close()

i = 0
while i <=504:
    i = i + 1
    mrt = (dataList[i][3],dataList[i][4])
    name = dataList[i][1]
    folium.Circle(
        location=mrt,
        radius=100,
        popup=parse_zhch(name),
        color='blue',
        fill=False,
        fill_color='blue',
        fill_opacity=0.05
    ).add_to(m)


# Starbucks(GREEN)
csvFileToRead = open('C:\\Users\\NB\\notebooks\\starbucklist.csv', 'r', encoding='utf-8_sig')
csvDataToRead = csv.reader(csvFileToRead)
dataList = list(csvDataToRead)
csvFileToRead.close()
i = 0
while i <=498:
    i = i + 1
    mrt = (dataList[i][5],dataList[i][6])
    name = dataList[i][1]
    folium.Circle(
        location=mrt,
        radius=100,
        popup=parse_zhch(name),
        color='green',
        fill=False,
        fill_color='green',
        fill_opacity=0.05,
    ).add_to(m)
m
```
