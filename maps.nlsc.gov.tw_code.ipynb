## 爬取目標地周邊設施資料
### 以捷運大安站為例(121.543551,25.032943)

```python
import requests 
import pandas as pd
import csv
```

```python
# 爬下目標地周圍1000公尺內的學校
# markclass:學校、醫療、嫌惡設施分類填1，工商分類填2
# mark可填入以下條件：大專院校,中學,小學,職訓中心,幼兒園,特殊學校,圖書館,博物館,資料及陳列館,文化中心,社教館,美術館,醫學中心醫院,衛生所,工廠,商號,公司
# radius半徑
# polygonBox格式:經度在前，緯度在後，如 121.543603,25.033157
    
post_params = {
    'marktype': 'Circle',
    'markclass': '1',
    'mark': '大專院校,中學,小學,幼兒園,特殊學校',
    'radius': 1000,
    'polygonBox': (121.543551,25.032943)
}

response_json = requests.post('https://maps.nlsc.gov.tw/T09/pro/MarkCluster_markpos.action', data = post_params).json()
result = pd.DataFrame.from_dict(response_json['Qrygroup1'])
result
```
